```python
TAG = df['headline'][0].split(':')[0]
tags = [x.lower() for x in TAG.split(' ')]
tags
```

```python
TAG
tags[0]
```

```python
headline_col = []
for line in df['headline']:
    print(line)
    for tag in tags:
        print(line.replace(TAG, tag.title()))
        headline_col.append(line.replace(TAG, tag))
```

```python
headline_col
```

```python
id_col = [x for x in range(1, len(headline_col)+1)]
```

```python
id_col[-1]
```

```python
print('id_col size = {0}, headline_col size = {1}'.format(len(id_col), len(headline_col)))
```

```python
newDf = {'id': id_col, 'headline': headline_col}
```

```python
Xdf = pd.DataFrame(data=newDf)
Xdf = Xdf.set_index('id')
```

```python
Xdf.to_csv('headlines.csv')
```
